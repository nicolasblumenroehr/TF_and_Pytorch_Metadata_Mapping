```python
import tensorflow_datasets as tfds
import glob
import skimage.io as io
import os
import random
import json
import pandas as pd
from typing import Any, ClassVar, Dict, Iterable, List, Optional, Type, Union
from tensorflow_datasets.core import utils
import tensorflow as tf
import inspect
from ipynb.fs.full.schemaExtractor import Extractor

class Custom_Dataset(tfds.core.GeneratorBasedBuilder):
    """DatasetBuilder for my_dataset dataset."""
    
    VERSION = tfds.core.Version('1.0.0')
    
    RELEASE_NOTES = {
      '1.0.0': 'Initial release.',
    }
    
    def __init__(self,
        *,
        data_dir: Optional[utils.PathLike] = None,
        config: Union[None, str, tfds.core.BuilderConfig] = None,
        version: Union[None, str, utils.Version] = None,       
        ):
        
        super().__init__()
    
    def _info(self):
        
        metadata=open(os.path.join((os.path.expanduser("~/temp/")),[f for f in os.listdir(os.path.expanduser("~/temp/")) if f.endswith('.json')][0]))
        jsonmetadata = json.load(metadata)
        digitsMetadata = Extractor(jsonmetadata)
        digitsMetadata.extractSchema()
        
        self.homepage = digitsMetadata.homepage
        self.dataset_name = digitsMetadata.name
        self.format = digitsMetadata.format
        
        return tfds.core.DatasetInfo(
        builder = self,
            
            description = digitsMetadata.description,
            features = tfds.features.FeaturesDict({
                'image': tfds.features.Image()
            }),
            homepage = digitsMetadata.homepage,
            metadata = tfds.core.MetadataDict(
                digitsMetadata.metadata),
            license = digitsMetadata.license,
            disable_shuffling=True,
            citation = digitsMetadata.citation
            
        )
        
    def _split_generators(self, dl_manager=tfds.download.DownloadManager):
        
        files = dl_manager.download_and_extract(url_or_urls=self.homepage)
        path = str(files) + "/" + self.dataset_name + "/"
        
        return [
                tfds.core.SplitGenerator(
                    name=tfds.Split.TRAIN,
                    gen_kwargs=dict(path=path)),
            ]
    
    def _generate_examples(self, path):
        
        "Generator of examples for each split. The content must correspond to the features defined in _info()"
        for img_path in glob.glob(path+"*"+self.format):
            yield img_path, {
                'image': img_path
                
          }
```
